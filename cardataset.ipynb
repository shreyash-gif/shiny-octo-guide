```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
car_data = pd.read_csv("data.csv")
car_data.head(5)
```

HEAD AND TAIL 

```python
car_data.tail(5)
```

DUPLICATES PRINTING

```python

df=pd.DataFrame(car_data,columns = ['Make','Model','Year','Engine Fuel Type','Engine HP','Engine Cylinders','Transmission Type','Driven_Wheels','Number of Doors','Vehicle Size','Vehicle Style','city mpg','highway MPG','Popularity','MSRP'])
duplicate = df[df.duplicated()]
duplicate
```

REMOVING DUPLICATES AND TOTALLING IT

```python
display(df.drop_duplicates())
total = df.duplicated().sum()
print(total)
```

```python
print("Maximum price of Car is ",car_data["MSRP"].max())
print("Minimum price of Car is ",car_data["MSRP"].min())
print("Average price of Car is ",car_data["MSRP"].mean())
```

SCATTERPLOT OF ENGINE HP Vs MSRP

```python
sns.scatterplot(data=car_data,x="Engine HP" ,y="MSRP")
```

HISTOGRAM OF COUNTING THE NUMBER OF DIFFERENT CARS

```python
plt.hist(data=car_data, x='Make',bottom=11.100)
ticks = np.arange(0, 50, 5.5) 
plt.xticks(ticks)
plt.xlabel('Make')
plt.ylabel('Count')
```

```python
print("BMW cars sales is highest in year",car_data[car_data.Make=="BMW"]["Year"].max())
```

**# FINDING RELATION BETWEEN NUMBERS OF CARS Vs PRICE**
It is almost inversly proportional after around 20,000 price

```python
plt.hist(df.MSRP,range=[0,100000])
plt.xlabel("Price")
plt.ylabel("Number of Cars")
plt.show()
```

# **Inversely Proportional Relation Between HighWay MPG and Price of Car**

```python
plt.scatter(df["highway MPG"],df["MSRP"])
plt.xlabel("Highway MPG")
plt.ylabel("Price of Car")
plt.show()
```
